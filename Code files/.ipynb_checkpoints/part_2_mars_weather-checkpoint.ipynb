## Deliverable 2: Scrape and Analyze Mars Weather Data

```python
# Import relevant libraries
from splinter import Browser
from bs4 import BeautifulSoup as soup
import matplotlib.pyplot as plt
import pandas as pd
```

```python
browser = Browser('chrome')
```

### Step 1: Visit the Website

Use automated browsing to visit the [Mars Temperature Data Site](https://static.bc-edx.com/data/web/mars_facts/temperature.html). Inspect the page to identify which elements to scrape.

```python
# Visit the website
# https://static.bc-edx.com/data/web/mars_facts/temperature.html
url = "https://static.bc-edx.com/data/web/mars_facts/temperature.html"
browser.visit(url)
```

### Step 2: Scrape the Table

Create a Beautiful Soup object and use it to scrape the data in the HTML table.

Note that this can also be achieved by using the Pandas `read_html` function. However, use Beautiful Soup here to continue sharpening your web scraping skills.

```python
# Create a Beautiful Soup Object
html = browser.html
mars_wx_soup = soup(html, 'html.parser')
```

```python
# Extract all rows of data 
table = mars_wx_soup.find('table', class_='table')
rows = table.find_all('tr')
```

### Step 3: Store the Data

Assemble the scraped data into a Pandas DataFrame. The columns should have the same headings as the table on the website. Hereâ€™s an explanation of the column headings:

* `id`: the identification number of a single transmission from the Curiosity rover
* `terrestrial_date`: the date on Earth
* `sol`: the number of elapsed sols (Martian days) since Curiosity landed on Mars
* `ls`: the solar longitude
* `month`: the Martian month
* `min_temp`: the minimum temperature, in Celsius, of a single Martian day (sol)
* `pressure`: The atmospheric pressure at Curiosity's location

```python
# Create an empty list
data = []

# Loop through the scraped data to create a list of rows (Loop through each row in table)
for row in rows:
    # Create empty list for this row's data
    row_data = []
    # Find all 'th' and 'td' elements in row
    cells = row.find_all(['th', 'td'])
    # Loop through each cell and extract text
    for cell in cells:
        obs = cell.text.strip()  # Get the text and remove any whitespace
        row_data.append(obs) # Add the text to the row_data list
    # Add the row's data to the data list
    data.append(row_data)

# Identify first row as header; remaining rows contain data
headers = data[0]  
data = data[1:]    

################################################
# # Alternative solution: loop over headers and separately loop over data rows, rather than use nested loops 

# # First <tr> contains table headers -- Extract headers
# header_row = table.find('tr')
# headers = []
# for th in header_row.find_all('th'):  
#     headers.append(th.text.strip())  

# # Remaining <tr> have class 'data-row' -- Extract data rows
# data_rows = table.find_all('tr', class_='data-row')
# data = []
# for row in data_rows:
#     row_data = []  
#     for td in row.find_all('td'):  
#         row_data.append(td.text.strip())  
#     data.append(row_data)
```

```python
# Create a Pandas DataFrame by using the list of rows and a list of the column names
df = pd.DataFrame(data, columns=headers)
```

```python
# Confirm DataFrame was created successfully -- Display the DataFrame
df.head()
```

### Step 4: Prepare Data for Analysis

Examine the data types that are currently associated with each column. If necessary, cast (or convert) the data to the appropriate `datetime`, `int`, or `float` data types.

  > **Hint** You can use the Pandas `astype` and `to_datetime` methods to accomplish this task.


```python
# Examine data type of each column
df.dtypes
```

```python
df['terrestrial_date'] = pd.to_datetime(df['terrestrial_date'])
df['sol'] = df['sol'].astype(int)
df['ls'] = df['ls'].astype(int)
df['month'] = df['month'].astype(int)
df['min_temp'] = df['min_temp'].astype(float)
df['pressure'] = df['pressure'].astype(float)
```

```python
# Confirm type changes were successful by examining data types again
df.dtypes
```

### Step 5: Analyze the Data

Analyze your dataset by using Pandas functions to answer the following questions:

1. How many months exist on Mars?
2. How many Martian (and not Earth) days worth of data exist in the scraped dataset?
3. What are the coldest and the warmest months on Mars (at the location of Curiosity)? To answer this question:
    * Find the average the minimum daily temperature for all of the months.
    * Plot the results as a bar chart.
4. Which months have the lowest and the highest atmospheric pressure on Mars? To answer this question:
    * Find the average the daily atmospheric pressure of all the months.
    * Plot the results as a bar chart.
5. About how many terrestrial (Earth) days exist in a Martian year? To answer this question:
    * Consider how many days elapse on Earth in the time that Mars circles the Sun once.
    * Visually estimate the result by plotting the daily minimum temperature.


```python
# 1. How many months are there on Mars?
num_months = df['month'].max()
num_months
```

```python
# 2. How many sols (Martian days) worth of data are there?
len(df)
```

```python
# 3. What is the average low temperature by month
ave_low_by_month = df.groupby('month')['min_temp'].mean()
# or: ave_low_by_month = df['min_temp'].groupby(df['month']).mean() 
ave_low_by_month
```

```python
# Plot the average minimum temperature by month

plt.bar(ave_low_by_month.index, ave_low_by_month.values, color='steelblue', width=0.5)

# Add labels and title
plt.xlabel('Month')
plt.ylabel('Temperature in Celsius (Â°C)')
plt.title('Average Minimum Temperature by Month')
plt.xticks(ave_low_by_month.index)  # Ensure all months are labeled

# Show the plot
plt.show()
```

```python
# Identify the coldest and hottest months in Curiosity's location by sorting the previous graph

# Sort data from highest to lowest
sorted_data = ave_low_by_month.sort_values(ascending=True)

# Plot the bar graph with sorted data
plt.bar(range(len(sorted_data)), sorted_data.values, color='steelblue', width=0.5)

# Add labels and title
plt.xlabel('Month')
plt.ylabel('Average Low Temperature (Â°C)')
plt.title('Average Low Temperature by Month (Sorted)')
plt.xticks(range(len(sorted_data)), sorted_data.index)  

# Show the plot
plt.show()
```

### Analysis: On average, the third Mars-month has the coldest (minimum) temperatures, while the eighth Mars-month has the warmest (by approximately 15 degrees C).

```python
# 4. Average pressure by Martian month
ave_pressure_by_month = df.groupby('month')['pressure'].mean()
ave_pressure_by_month
```

```python
# Plot the average pressure by month

plt.bar(ave_pressure_by_month.index, ave_pressure_by_month.values, color='steelblue', width=0.5)

# Add labels and title
plt.xlabel('Month')
plt.ylabel('Atmospheric Pressure')
plt.title('Average Pressure by Month')
plt.xticks(ave_pressure_by_month.index)  # Ensure all months are labeled

# Show the plot
plt.show()
```

### Analysis: On average, the sixth Mars-month has the lowest atmospheric pressure, while the ninth Mars-month has the greatest pressure.

```python
# 5. How many terrestrial (earth) days are there in a Martian year?

# Plot the daily minimum temperature of each observation in the data set - Plot a line graph
plt.plot(df.index, df['min_temp'], linestyle='-', color='steelblue')

# Add labels and title
plt.xlabel('Number of Terrestrial Days')
plt.ylabel('Minimum Temperature  (Â°C)')
plt.title('Minimum Temperature Over Time')

# Show the plot
plt.show()
```

### Analysis: There are 3 peaks, and 3 troughs, in the last graph, suggesting three Mars-years of data. The distance from the last two troughs is approxinately 1750-1100, or 650, terristrial (i.e., earth) days. Internet search indicates that a Mars year is equivalent to 687 earth days.
